```python
import numpy as np
from scipy.special import loggamma, factorial, gamma
```

```python
def poiss_likelihood(x: np.ndarray, theta: float) -> float:
    return np.exp(-theta * x.size) * theta ** x.sum() * (1 / factorial(x)).prod()
```

```python
def poiss_loglikelihood(x: np.ndarray, theta: float) -> float:
    return -x.size * theta + np.log(theta) * x.sum() - loggamma(x + 1).sum()
```

```python
def poiss_mle(x: np.ndarray) -> float:
    return x.mean()
```

```python
def poiss_av(theta: float) -> float:
    return theta
```

```python
def gamma_likelihood(x: np.ndarray, theta: float, beta: float) -> float:
    return theta ** (x.size * beta) / gamma(beta) ** x.size * np.exp(-theta * x.sum()) * (x ** (beta - 1)).prod()
```

```python
def gamma_loglikelihood(x: np.ndarray, theta: float, beta: float) -> float:
    return x.size * beta * np.log(theta) - x.size * loggamma(beta) - theta * x.sum() + (beta - 1) * np.log(x).sum()
```

```python
def gamma_mle(x: np.ndarray, beta: float) -> float:
    return beta/ x.mean()
```

```python
def gamma_av(theta: float, beta: float) -> float:
    return theta ** 2 / beta
```
