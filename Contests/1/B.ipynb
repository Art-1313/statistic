## B. Ближайшие соседи Ph@DS

```python
import numpy as np
import typing as tp

A = np.array([[0, 0, 0], [4, 5, 6], [7, 8, 9], [0, 0, 0]])
B = np.array([[4, 5, 6], [7, 8, 9], [1, 2, 3]])
k = 2

def find_nearest_points(A: np.ndarray, B: np.ndarray, k: int) -> np.ndarray:
    return np.ones((B.shape[0], k)) + np.argsort(((np.full((A.shape[0], B.shape[0], B.shape[1]), B) - np.full((B.shape[0], A.shape[0], A.shape[1]), A).transpose(1, 0, 2)) ** 2).sum(axis=2).transpose(), axis=1)[:,:k]

find_nearest_points(A, B, k)
```
