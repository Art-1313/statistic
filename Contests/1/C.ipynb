## C. Подотрезок с максимальной суммой Ph@DS

```python
import numpy as np
import typing as tp

A = np.array([5, 3, -6, 1])
k = 2

def find_max_sum_segment(array: tp.List[int], k: int) -> int:
    return (np.triu(np.full((array.__len__() - k + 1, array.__len__()), array)) - np.triu(np.full((array.__len__() - k + 1, array.__len__()), array), k)).sum(axis=1).max()

find_max_sum_segment(A, k)
```
